```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar grayscale
img = cv2.imread('gambar_2.png', 0)
if img is None:
    raise ValueError(" Gambar tidak ditemukan! Pastikan nama file benar dan sudah di-upload.")

plt.figure(figsize=(5,5))
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli (Dengan Noise)')
plt.axis('off')
plt.show()
```

```python
def fourier_transform(img):
    F = np.fft.fft2(img)
    Fshift = np.fft.fftshift(F)
    return Fshift

def inverse_fourier(Fshift):
    F_ishift = np.fft.ifftshift(Fshift)
    img_back = np.fft.ifft2(F_ishift)
    return np.abs(img_back)
```

LOW PASS FILTER

```python
rows, cols = img.shape
crow, ccol = rows//2, cols//2
Fshift = fourier_transform(img)

# Buat mask lingkaran (low pass)
mask = np.zeros((rows, cols), np.uint8)
r = 30  # radius filter
cv2.circle(mask, (ccol, crow), r, 1, -1)

# Terapkan filter
F_low = Fshift * mask
img_low = inverse_fourier(F_low)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.imshow(mask, cmap='gray'); plt.title('Low Pass Mask')
plt.subplot(1,2,2); plt.imshow(img_low, cmap='gray'); plt.title('Hasil Low Pass Filter')
plt.show()
```

HIGH PASS FILTER

```python
mask = np.ones((rows, cols), np.uint8)
r = 30
cv2.circle(mask, (ccol, crow), r, 0, -1)

F_high = Fshift * mask
img_high = inverse_fourier(F_high)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.imshow(mask, cmap='gray'); plt.title('High Pass Mask')
plt.subplot(1,2,2); plt.imshow(img_high, cmap='gray'); plt.title('Hasil High Pass Filter')
plt.show()
```

NOCTH FILTER

```python
mask = np.ones((rows, cols), np.uint8)

# Contoh notch di sekitar titik-titik tertentu
notches = [(crow+30, ccol+30), (crow-30, ccol-30)]
for (y, x) in notches:
    cv2.circle(mask, (x, y), 10, 0, -1)

F_notch = Fshift * mask
img_notch = inverse_fourier(F_notch)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.imshow(mask, cmap='gray'); plt.title('Notch Filter Mask')
plt.subplot(1,2,2); plt.imshow(img_notch, cmap='gray'); plt.title('Hasil Notch Filter')
plt.show()
```

MEDIAN FILTER

```python
img_median = cv2.medianBlur(img, 3)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Asli (Dengan Noise)')
plt.subplot(1,2,2); plt.imshow(img_median, cmap='gray'); plt.title('Hasil Median Filter')
plt.show()
```

GAUSSIAN FILTER

```python
img_gaussian = cv2.GaussianBlur(img, (5,5), 1)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Asli (Dengan Noise)')
plt.subplot(1,2,2); plt.imshow(img_gaussian, cmap='gray'); plt.title('Hasil Gaussian Filter')
plt.show()
```
